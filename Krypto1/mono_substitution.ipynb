```python
from Decrypter import Decrypter
import matplotlib.pyplot as plt

english_freq =    {'E': 13.0, 'T': 9.10, 'A': 8.2, 'O': 7.5, 'I': 6.97, 'N': 6.75, 'S': 6.33,
                  'H': 6.09, 'R': 5.99, 'D': 4.25, 'L': 4.03, 'C': 2.76, 'U': 2.78, 'M': 2.4,
                  'W': 2.4, 'F': 2.02, 'G': 2.36, 'Y': 2.02, 'P': 1.93, 'B': 1.49, 'V': 0.98,
                  'K': 0.77, 'J': 0.15, 'X': 0.15, 'Q': 0.09, 'Z': 0.07}

decrypter = Decrypter("chiffrat.txt")
print(decrypter.chiffrat)
```

```python
letter_frequency = decrypter.calculate_r_frequency(decrypter.chiffrat)
```

```python
# Extracting the keys and values from the dictionaries
keys1 = list(english_freq.keys())
values1 = list(english_freq.values())
keys2 = list(letter_frequency.keys())
values2 = list(letter_frequency.values())

# Creating a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,6))

# Plotting the first graph
bars1 = ax1.bar(keys1, values1, label='Häufigkeit Engllish', color='#0066ff')
ax1.set_xlabel('Char')
ax1.set_ylabel('Buchstaben Häufigkeit im chiffrat')
ax1.legend()

# Add values to the first graph
for bar1 in bars1:
    height = bar1.get_height()
    ax1.text(bar1.get_x() + bar1.get_width() / 2., height, '{:.1f}'.format(height),
            ha='center', va='bottom')

# Plotting the second graph
bars2 = ax2.bar(keys2, values2, label='Relative Häufigkeit der Chiffrat %', color='red')
ax2.set_xlabel('Char')
ax2.set_ylabel('%')
ax2.legend()

# Add values to the second graph
for bar2 in bars2:
    height = bar2.get_height()
    ax2.text(bar2.get_x() + bar2.get_width() / 2., height, '{:.1f}'.format(height),
            ha='center', va='bottom')

# Displaying the plot
plt.show()
```
