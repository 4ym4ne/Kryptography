```python
from Decrypter import Decrypter
import matplotlib.pyplot as plt

decrypter = Decrypter("chiffrat2.txt")
print(decrypter.chiffrat)
```

```python
absolute_frequency = dict(sorted(decrypter.calculate_a_frequency(decrypter.chiffrat).items(), key=lambda x: -x[1]))
relative_frequency = decrypter.calculate_r_frequency(decrypter.chiffrat)

print(absolute_frequency)
print(relative_frequency)
```

```python

# Extracting the keys and values from the dictionaries
keys1 = list(absolute_frequency.keys())
values1 = list(absolute_frequency.values())
keys2 = list(relative_frequency.keys())
values2 = list(relative_frequency.values())

# Creating a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plotting the first graph
ax1.bar(keys1, values1, label='Absolute Häufigkeit', color='#0066ff')
ax1.set_xlabel('Char')
ax1.set_ylabel('Anzahl der buchstaben im chiffrat')
ax1.legend()

# Plotting the second graph
ax2.bar(keys2, values2, label='Relative Häufigkeit %', color='red')
ax2.set_xlabel('Char')
ax2.set_ylabel('%')
ax2.legend()

# Displaying the plot
plt.show()
```
